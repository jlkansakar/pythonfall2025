# Generators
You can create generators in 3 ways
* Itterator classes
* Generator functions
* Genreator expression

A gernerator is an object that follows the iterator protocol.
* Should implement 
    * \_\_iter\_\_ 
    * \_\_next\_\_
    * raise Stopiteration

this will make it able to work in code like a for loop

```
    for i in range(10):
        print(i)

    for i in [1, 2, 3, 4, 5]:
        print(i)
```

It will react with the build in functions:
* iter()
* next()

## 1. Iterator classes

````python
class compute:
    def __init__(self, num):
        self.num = num      
    def __iter__(self):
        self.start = 0
        return self
    def __next__(self):
        rv = self.start
        self.start += 1
        if self.start > self.num:
            raise StopIteration('You reached the end!')
        return rv
````

````python
it = iter(compute(3))
````

````python
next(it)
````

````python
for i in compute(2):
    print(i)
````

## 2. Generator functions

````python
def compute(num):
    for i in range(num):
        yield i
````

````python
it = compute(10)
````

````python
next(it)
````

## 3. Generator expression

````python
it = ( i for i in range(10)  )
````

````python
next(it)
````
